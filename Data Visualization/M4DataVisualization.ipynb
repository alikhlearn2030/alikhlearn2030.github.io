# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting with Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
# your code goes here
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'].dropna(), vert=False, patch_artist=True)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
df_clean = df.dropna(subset=['Age', 'WorkWeekHrs'])
```

```python
print(df_clean[['Age', 'WorkWeekHrs']].isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['Age'], df_clean['WorkWeekHrs'], alpha=0.5, edgecolor='k')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
# Drop rows with missing values in the columns of interest
df_clean = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

bubble_size = df_clean['Age'] / df_clean['Age'].max() * 1000  # Scale factor for bubble sizes
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df_clean['WorkWeekHrs'], df_clean['CodeRevHrs'], s=bubble_size, alpha=0.5, edgecolor='k')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
"""
df = pd.read_sql_query(QUERY,conn)
print(df)

```

```python
# your code goes here
plt.figure(figsize=(10, 7))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140, colors=plt.get_cmap('tab10').colors)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

```python
# SQL query to get the count of each language respondents wish to learn
conn = sqlite3.connect('m4_survey_data.sqlite')

# Query to get the names of all tables
tables_query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(tables_query, conn)

# Close the connection
conn.close()

# Display the table names
print(tables)
```

```python
# Connect to the SQLite database
conn = sqlite3.connect('m4_survey_data.sqlite')

# SQL query to get the count of each language respondents wish to learn
QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) AS count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY count DESC
"""

# Execute the query and store results in a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Display the DataFrame to understand its structure
print(df.head())

# Determine the rank of Python
if 'Python' in df['LanguageDesireNextYear'].values:
    python_rank = df[df['LanguageDesireNextYear'] == 'Python'].index[0] + 1
    print(f'Python is ranked {python_rank} in the list of most popular languages respondents wish to learn next year.')
else:
    print('Python is not listed among the top languages respondents wish to learn next year.')
```

```python
conn = sqlite3.connect('m4_survey_data.sqlite')

# SQL query to count respondents working with SQL
QUERY = """
SELECT COUNT(*) AS count
FROM LanguageDesireNextYear
WHERE LanguageDesireNextYear LIKE '%SQL%'
"""

# Execute the query and store results in a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the connection
conn.close()

# Extract and print the count
count_sql_users = df['count'].iloc[0]
print(f'Number of respondents working with SQL: {count_sql_users}')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Filter data for age group 30 to 35
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY,conn)
df_median = df.groupby('Age').median()

df_median .plot(kind='bar', figsize=(10, 6), stacked=True)

plt.title('Stacked Bar Chart of Median WorkWeekHrs and CodeRevHrs for Those Age 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age,ConvertedComp
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
# your code goes here
df_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate median ConvertedComp for each age
median_comp_by_age = df_filtered.groupby('Age')['ConvertedComp'].median()

# Plotting the median ConvertedComp by Age
plt.figure(figsize=(12, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='b')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(median_comp_by_age.index)  # Ensure each age is a tick
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master

"""
df = pd.read_sql_query(QUERY,conn)
print(df)
```

```python
# Count occurrences of each value in 'MainBranch'
branch_counts = df['MainBranch'].value_counts()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
branch_counts.plot(kind='barh', color='skyblue', edgecolor='k')
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x')

plt.show()
```
